```python
sns.heatmap(df.corr(),cmap='coolwarm')
plt.title('Correlation Of Columns')
```

```python
#Price distribution
sns.displot(x = 'price', 
            data = df)
```

```python
#Relationship between the price and the amount of pages in a book
sns.jointplot(x = 'price', 
              y = 'pages', 
              data = df, 
              kind = 'scatter')
```

```python
#Relationship between the average review and the amount of pages in a book
sns.jointplot(x = 'pages', 
              y = 'avg_reviews', 
              data = df, 
              kind = 'scatter')
```

```python
#Relationship between the average review and price in a book
sns.jointplot(x = 'price',
              y = 'avg_reviews',
              data = df, 
              kind = 'scatter')
```

```python
#Average reviews
df.sort_values(by = 'avg_reviews',
              ascending = False)
```

```python
#Distribution Of Reviews
sns.displot(df['avg_reviews'],
            bins = 50)
```
